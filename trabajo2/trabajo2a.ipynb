## Punto 1 - Intervalo de confianza

Construir en Python un IC para al 95% de confianza necesario para responder: 쯇uede afirmarse que el resultado medio obtenido por los colegios en Lectura es superior a 45 puntos? Justifique su respuesta.

### Soluci칩n

Sea **$X_i$** la VA relacionada con el promedio obtenido en la prueba de lectura critica por el i-esimo colegio 

A partir de la tabla podemos obtener los siguientes valores:

```python
# Datos del problema
data_prom_lectura = df.PROMLECT.copy()


# Calculo de la media y la desviacion estandar muestral
n = data_prom_lectura.shape[0]
X = data_prom_lectura.mean()
S = data_prom_lectura.std()
# Despliegue de los resultados
print(f"n = {n}")
print(f"X = {X:.5f}")
print(f"S = {S:.5f}")
```

Como la poblacion no es normal, el numero de datos que esta tiene es de mas de 30 y la varianza es desconocida, usamos la siguiente formula para el intervalo de confianza:

$$
IC = \overline{X}\pm Z_{\alpha/2}\frac{S}{\sqrt{n}}
$$

Inicialmente vamos a calcular para $\alpha = 0.05$:

$
Z_{\alpha/2} = Z_{0.025}  
$

```python
NC = 95 # Nivel de confianza
alpha = 1 - NC/100 

Z = stats.norm().ppf(1 - alpha/2)
print(f"Z = {Z:.5f}")

EBM = Z*S/np.sqrt(n)
print(f"EBM = {EBM:.5f}")

IC_left = X - EBM 
IC_right = X + EBM
print(f"({X:.5f} - {EBM:.5f},{X:.5f} + {EBM:.5f})= ({IC_left:.5f},{IC_right:.5f})")
```

**Respuesta numerica**: Como el intervalo en el cual se encuentra la media en los para lectura esta en el rango $(49.99748,50.64557)$, se puede afirmar con un nivel de confianza del 95% que el resultado promedio en la prueba de lectura es superior a 45 puntos.

Como el l칤mite inferior del intervalo de confianza es 49.997, que es mayor que 45, podemos afirmar con un 95% de confianza que el promedio de los resultados en Lectura es superior a 45 puntos. El valor de 45 puntos no est치 dentro del intervalo de confianza. Dado que todo el intervalo est치 por encima de 45 puntos, podemos concluir que es muy poco probable que el promedio verdadero sea igual o menor a 45 puntos. Por lo tanto, podemos afirmar que el promedio es superior a 45 puntos con un alto nivel de confianza.

## Punto 2 - Prueba de hipotesis

쯉e puede afirmar que el resultado medio obtenido por los colegios en Matem치ticas es mayor en los Privados que en los P칰blicos? Justifique su respuesta mediante una prueba de hip칩tesis con una significancia del 4%.

```python
# Informacion de la prueba de matematicas 
# publicos
df_publico = df[df['NATURALEZA'] == 'OFICIAL']
print(df_publico.shape)
# privados
df_privado = df[df['NATURALEZA'] == 'NO OFICIAL']
print(df_privado.shape)

# Nivel de significancia
alpha = 0.04 # 4%
```

```python
notas_mat_privado = df_privado.PROMMATE
notas_mat_publico = df_publico.PROMMATE
```

Se tienen las siguientes VA:
* Sea **$X_i$** la VA relacionada con el promedio obtenido en la prueba de matematicas por el i-esimo colegio privado
* Sea **$Y_i$** la VA relacionada con el promedio obtenido en la prueba de matematicas por el j-esimo colegio publico

Tenemos los siguientes datos por colegio:
* **Colegio privado**:
  * $1466$
* **Colegio publico**:

```python
# Colegio privado
nx = notas_mat_privado.shape[0]
X = notas_mat_privado.mean()
Sx = notas_mat_privado.std()
print(f"nx = {nx}")
print(f"X = {X:.5f}")
print(f"Sx = {Sx:.5f}")
```

```python
# Colegio publico
ny = notas_mat_publico.shape[0]
Y = notas_mat_publico.mean()
Sy = notas_mat_publico.std()
print(f"ny = {ny}")
print(f"Y = {Y:.5f}")
print(f"Sy = {Sy:.5f}")
```

Hipotesis: Enunciado: $\mu_x > \mu_y$ es decir $\mu_x - \mu_y > 0$

* $H_o: \mu_x - \mu_y \leq 0$
* $H_a: \mu_x - \mu_y > 0$

Calculo del estadistico de prueba

```python
d0 = 0
Zc = (X - Y - d0) / ((Sx**2 / nx + Sy**2 / ny)**0.5)
print(Zc)
```

```python
VP = 1 - stats.norm.cdf(Zc)
print(VP)
```

```python
# Paso 4 - Si VP <= a, se rechaza la hipotesis nula, para un a = 0.05
print(VP >= alpha)
```

Como el valor de  洧녾 z es mucho mayor que los valores cr칤ticos t칤picos para un nivel de significancia de 4% (que es aproximadamente 1.75 para una prueba unilateral), podemos rechazar la hip칩tesis nula con confianza, lo que confirma que el resultado medio en Matem치ticas es mayor en los colegios privados que en los p칰blicos.
